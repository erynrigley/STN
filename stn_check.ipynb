```python
import graphviz
from graphviz import Digraph
import sys
import numpy as np

class STN:

    ## useful maximum upper bound
    inf = sys.float_info.max
    
    ## list of nodes and adjacency list
    def __init__(self):
        STN.nodes = []
        STN.adj_list = {}
        STN.child_list = []
    
                #---------------------#
                # 2.2.Floyd-Warshall  #
                #---------------------#
    
    def check_consistency(self):   
        for x in range(len(stn.child_list) + 1): #create matrix with values
            dist = np.zeros(shape=(x,x))

        #for every (parent) node in tree: 
        for u in stn.adj_list.keys(): 
            dic = stn.adj_list[u]
            #for every (connecting) node:
            for v in dic:
                #fill in the value of that connection
                dist[stn.child_list.index(u)][stn.child_list.index(v)] = dic[v]

        for k in range(len(stn.nodes)):                            #'meat' of algorithm
            for i in range(len(stn.nodes)):                        #for every triple
                for j in range(len(stn.nodes)):
                    if dist[i][j] > (dist[i][k] + dist[k][j]):
                        dist[i][j] = (dist[i][k] + dist[k][j])     #find minimal cost path between them
                        if 1 == j and dist[i][j] < 0:              #and if negative self loop, 
                            return False                           #strategy = inconsistent
        return True, dist                                          #also returning distance matrix, as i'll
                                                                   #need it later on
                #---------------------#
                # 2.3.  Dot-File      #
                #---------------------#
                
    def print_stn(self):
        dist = stn.check_consistency()[1]
        dot = Digraph()
        
        #add nodes
        for node in stn.nodes:
            name = node
            dot.node(name)
        
        #add edges
        # #for every (parent) node in tree: 
        for u in stn.adj_list.keys(): 
            dic = stn.adj_list[u]
            #for every (connecting) node:
            for v in dic:
                #create edge + time label
                label_ = str(dist[stn.child_list.index(u)][stn.child_list.index(v)])
                dot.edge(u, v, label = label_)

        return dot
    
                #---------------------#
                # 2.4. Make-Minimal   #
                #---------------------#
    
    def make_minimal(self):
        for k in stn.child_list:             #for every triple of nodes
            for i in stn.child_list: 
                for j in stn.child_list:

                    #kdic = stn.adj_list[k] #
                    #idic = stn.adj_list[i] #their corresponding dicionaries
                    #jdic = stn.adj_list[j]
                    
                    if i in stn.nodes and k in stn.nodes:    #1. if i->j, and i ->k and k ->j
                        if k in stn.adj_list[i] and j in stn.adj_list[k]:
                            if j in stn.adj_list[i]:
                                                            # and-
                                                            
                                                             # [2. dist[i->j] <0 and dist[i ->k] <0
                                                             #3. or dist[i -> j] >/ 0 and dist[k->j] >/0 ]
                                if ((dist[i][j] < 0) and (dist[i][k] < 0)) or ((dist[i][j] <= 0) and (dist[k][j] <= 0)):
                                    del (stn.adj_list[i])[j]     #and delete
```

## 2.1 Parsing STN from file 

```python
## open the DOT file
filename = "match_plan_1.dot"
with open(filename, "r") as dotfile:

    stn = STN()
    
    for line in dotfile:
        ## read edge
        
        if "->" in line:
            
                #---------------------#
                # 2.1. Parsing STN    #
                #---------------------#
            
            ## save nodes in list stn.nodes 
            
            node = line[:(line.find("->"))]
            node = node.strip('\t')
            node = node.strip()
            
            stn.nodes.append(node)
            
                        
            #save connecting nodes + times in stn.adj_list + child_list
            child = line[(line.find("->")+3):(line.find("[")) ]
            child = child.strip()

            stn.child_list.append(child)
            
            time = line[ (line.find("=")+2):(line.find("]") -2 )]
            
            ## initialise adjacency list stn.adj_list
            if node in stn.adj_list.keys():
                stn.adj_list[node].update({child:time})
            else:
                stn.adj_list[node] = {child:time}
                
stn.nodes = list(dict.fromkeys(stn.nodes)) #removes duplicates from stn.nodes
stn.nodes = [x.replace('\t', '').replace('\n', '').replace(" ", "") for x in stn.nodes] #remove \ts + |ns from stn.nodes


#main function
if stn.check_consistency():
    dot = stn.print_stn()
    dot.render('plan_original.gv', view=True) #first plan
    stn.make_minimal()
    dot = stn.print_stn()
    dot.render('plan_minimal.gv', view=True) #minimal plan
    print('CONSISTENT')
else:
    print("NOT CONSISTENT")
```
